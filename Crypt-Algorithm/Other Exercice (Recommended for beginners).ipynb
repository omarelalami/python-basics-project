  Exercice 00 - Factorial

</div>


```python
def factortial(n):
    res = 1
    for index in range(1,n+1):
        res = index * res
    return res
```

```python
print(factortial(6))
```

```python
import time
def premier(n):
    cmt=0
    for c in range(1,int(n/2)+1):
       
        if n%c==0:
            cmt+=1
        if cmt>1:
            return False
            
        elif cmt<=1 and c==int(n/2):
            return True
print(premier(2))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2-8*np.log(x)
```

```python
x = np.linspace(0.5, 5, 450) 
y = f(x)
```

```python
plt.figure(figsize=(18,12))
plt.plot(x, y, label='f(x) = xÂ²-8.ln(x)', color='r')
plt.xlabel('x')
plt.ylabel('Y')
plt.title('Graphique de f(x)')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend()
```
